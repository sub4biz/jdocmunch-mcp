# Quickstart

This notebook walks through importing the library and running a basic parse.


## Setup

Install via pip and import the parser.


```python
from examplelib import parse
result = parse(open('server.log'))
print(len(result), 'events')
```

## Filtering events

Filter the parsed events by level. The events are dicts with `timestamp`, `level`, and `message` keys.


```python
errors = [e for e in result if e['level'] == 'ERROR']
for e in errors:
    print(e['timestamp'], e['message'])
```
